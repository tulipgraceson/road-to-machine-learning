# The Basics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ds=pd.read_csv("Position_Salaries.csv")
x=ds.iloc[:,1:-1].values
y=ds.iloc[:,-1].values

print(x)
print(y)
```

# Training the Decsion Tree Regression model on the whole dataset

```python
from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor(random_state=0)
reg.fit(x,y)
```

# Predicting the output

```python
print(reg.predict([[6.5]]))
y_pred=reg.predict(x)
print(y_pred)
```

**U CANT PLOT FOR DATASETS FOR MORE THAN ONE FEATURE AS ITS HAVING MORE THAN 2 DIMESNIONS**

# visualizing the output

```python
plt.scatter(x,y,color="red")
plt.plot(x,y_pred,color="blue")
```

*AS we can see this is not accurate and does not makes sense as they are split in groups*

# Visualizing Decision Tree Regression (higher resolution)

```python
x_grid=np.arange(min(x),max(x),0.1)
x_grid=x_grid.reshape(len(x_grid),1)
plt.scatter(x,y,color="red")
plt.plot(x_grid,reg.predict(x_grid),color="blue")
```
