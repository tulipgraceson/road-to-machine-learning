**IMPORTING AND ASSIGNING**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Position_Salaries.csv")
x=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
print(x)
print(y)
```

**Training the linear regression model on the whole dataset**

```python
from sklearn.linear_model import LinearRegression
rg=LinearRegression()
rg.fit(x,y)
y_pred=rg.predict(x)
print(y_pred)
```

**Training the polynomial regression model on whole dataser**

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=6)
x_poly=poly.fit_transform(x)
print(x_poly)
```

```python
poly_rg=LinearRegression()
poly_rg.fit(x_poly,y)
```

```python
y_poly_pred=poly_rg.predict(x_poly)
print(y_poly_pred)
```

**visuallizing linear regression model***

```python
plt.scatter(x,y,color="red")
plt.plot(x,y_pred,color="blue")
plt.title("linear regression model")
plt.xlabel("level")
plt.ylabel("salary")
plt.show()
```

**visuallizing polynomial regression model**

```python
plt.scatter(x,y,color="red")
#plt.scatter(x,y_poly_pred,color="green")
plt.plot(x,y_poly_pred,color="blue")
plt.title("poly regression model")
plt.xlabel("level")
plt.ylabel("salary")
```
